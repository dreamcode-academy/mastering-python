## Import necessary libraries

```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
```

#### Connect to the SQLite database

```python
conn = sqlite3.connect('weather_data.db')
```

### Perform analysis 

```python
df = pd.read_sql('SELECT * FROM weather_forecast', conn)
mean_temperature = np.mean(df['Temperature'])
median_wind_speed = np.median(df['Wind Speed'])

print(f"Mean Temperature: {mean_temperature}")
#print(f"Median Wind Speed: {median_wind_speed}")
```

### Plotting temperature

```python
plt.figure(figsize=(10,6))
plt.plot(df['Temperature'], label='Temperature')
plt.xlabel('Period')
plt.ylabel('Temperature Over Time')
plt.legend()
plt.show()

conn.close()
```
